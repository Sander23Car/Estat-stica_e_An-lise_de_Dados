# **DISTRIBUIÇÃO GAUSSIANA**

A distribuição Gaussiana, chamada na literatura também por distribuição normal, é um dos mais importantes exemplos de **distribuição contínua** de probabilidade.

## **Definição da distribuição gaussiana**

A função de densidade para esta distribuição é dada por

\begin{equation}
f(x) = P(X =x) = \frac{1}{\sigma \sqrt{2\pi}} \text{e}^{-(x-\mu)^2/2\sigma^2},
\end{equation}

onde $\mu$ é a média, $\sigma$ o desvio padrão e $\sigma^2$ a variância da distribuição. E $-\infty < x < \infty$.

A distribuição é chamada de normal porque todo esse o seu amassamento e alongamento foi especialmente organizado para dar a ela propriedades simples.

Para este exercício vamos utilizar a distribuição de Poisson dada no exercício 5, gerar dados com valores de $\lambda = 1, 5, 20, 100, 1000$
e comparar com a distribuição gaussiana considerando esta com média $\lambda$ e desvio padrão $\sqrt{\lambda}$.

# **CÓDIGOS**

```python
# Bibliotecas
import numpy as np
import scipy as sc
import scipy.special as scs
import math as mt
import matplotlib.pyplot as plt
```

```python
n = 100000

lamb_valores = [1, 5, 20, 100, 1000]

# Função Distribuição Gaussiana
def distgaussiana(x, media, sigma):
        return np.exp(-np.power(x - media, 2.) / (2 * np.power(sigma, 2.))) * (1/ (np.sqrt(2 * np.pi)*sigma))

for lamb in lamb_valores:
   p_distpoisson = np.random.poisson(lamb,n)
   x_list = np.linspace(lamb-3*mt.sqrt(lamb),lamb+3*mt.sqrt(lamb),1000)
   p_distgaussiana = [distgaussiana(a, lamb, mt.sqrt(lamb)) for a in x_list]

   # Plot dos gráficos
   plt.hist(p_distpoisson, label=f'Lambda={lamb}', color ='b', bins=100, density=True) # Distribuição de Poisson (Histograma -  cor azul)
   plt.plot(x_list, p_distgaussiana, color='red') # Distribuição Gaussiana (Curva - cor vermelha)
   plt.show()
```

## **CONCLUSÃO:**

Quanto maior for $n$ e $\lambda$ mais a distribuição de Poisson se aproxima para uma curva Gaussiana.
